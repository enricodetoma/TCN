```python
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def plot(fnames, labels, y_label):
    plt.clf()
    for idx, fname in enumerate(fnames):
        f = open(fname, 'r')
        x = []
        y = []
        i = 0
        for line in f.readlines():
            if i > 280:
                break
            tmp = line.split(',')
            if int(tmp[0]) == 1 or int(tmp[0]) % 10 == 0:
                i += 1
                x.append(int(tmp[0]))
                y.append(float(tmp[1]))
        print(len(x))
        f.close()
        plt.plot(x, y, label=labels[idx])
    plt.xlabel('Iteration', fontsize=15)
    plt.ylabel(y_label, fontsize=15)
    plt.legend(frameon=False, fontsize=12)
```

```python
fnames = ['./TCN/test_AUC.txt', './BasicDKT/test_AUC.txt']
labels = ['TCN', 'LSTM']
plot(fnames, labels, 'Test AUC')
```

```python
fnames = ['./TCN/test_acc.txt', './BasicDKT/test_acc.txt']
labels = ['TCN', 'LSTM']
plot(fnames, labels, 'Test Accuracy')
```

```python
fnames = ['./TCN/train_acc.txt', './BasicDKT/train_acc.txt', './TCN/test_acc.txt', './BasicDKT/test_acc.txt']
labels = ['TCN train', 'LSTM train', 'TCN test', 'LSTM test']
plot(fnames, labels, 'Train Accuracy')
```
